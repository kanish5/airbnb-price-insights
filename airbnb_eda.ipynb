```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv('airbnb_cleaned.csv')

# Show basic info
print(df.head())

# Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.grid(True)
plt.show()
# Price by Room Type
plt.figure(figsize=(8, 4))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Airbnb Price by Room Type')
plt.show()
# Average Price by Neighbourhood Group
avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(x=avg_price.index, y=avg_price.values)
plt.title('Average Airbnb Price by Neighbourhood Group')
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood Group')
plt.show()
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```
